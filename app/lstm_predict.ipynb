```python
prompt = "dawno temu czerwony kapturek poszedł do lasu i gdy szedł obok rzeki"
```

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 

from tqdm import tqdm
from keras.models import Model, load_model
from lstm import *
from keras import losses
from keras.preprocessing.text import tokenizer_from_json
import json 

with open("../datasets/words/books-bajki-raw-tokenizer_100000.json", "r") as f:
    data = json.load(f)
    tokenizer = tokenizer_from_json(data)


model: Model = load_model("../lstm_models/model_best_5_retrain.h5", custom_objects={"WeightedSCCE":  losses.SparseCategoricalCrossentropy})

for temperature in [0, 0.1, 0.2, 0.3, 0.5, 0.8, 1.0]:
    pr = PredictCallback(prompt, 100, LEN_MAX_LIMIT, tokenizer,'pre',  temperature, model)
    text = pr.generate_text()
    print(f" --- Temperature: {temperature} --- ")
    print(text)
```
